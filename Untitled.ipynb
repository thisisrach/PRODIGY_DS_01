Task 1: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Reading the data from csv file

data=pd.read_csv('world_population.csv')
```

```python
data.head()
```

```python
# Analysing the data with various plots and creating histogram

sns.barplot(x='Continent', y='2022 Population', data=data)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Continent', data=data,palette='colorblind')
```

```python
sns.histplot(x='Continent',data=data,kde=True,shrink=0.7)
```
